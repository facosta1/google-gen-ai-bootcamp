```python
# sql_engine: bigquery
# output_variable: df
# start _sql
_sql = """
CREATE OR REPLACE TABLE
  `alaska_dept_snow.alaska_snow_data_with_embeddings` AS
SELECT
  *
FROM
  ML.GENERATE_EMBEDDING( MODEL `alaska_dept_snow.embedding_model`,
    (
    SELECT
      string_field_0,
      string_field_1,
      CONCAT(string_field_0, string_field_1) AS content
    FROM
      `alaska_dept_snow.alaska_dept_snow_raw`) );
""" # end _sql
from google.colab.sql import bigquery as _bqsqlcell
df = _bqsqlcell.run(_sql)
df
```

```python
# sql_engine: bigquery
# output_variable: df
# start _sql
_sql = """
SELECT
  query.query,
  base.string_field_0,
  base.string_field_1
FROM
  VECTOR_SEARCH( TABLE `alaska_dept_snow.alaska_snow_data_with_embeddings`,
    'ml_generate_embedding_result',
    (
    SELECT
      text_embedding,
      content AS query
    FROM
      ML.GENERATE_TEXT_EMBEDDING( MODEL `alaska_dept_snow.embedding_model`,
        (
        SELECT
          'Weather' AS content)) ),
    top_k => 5,
    OPTIONS => '{"fraction_lists_to_search": 0.01}')
""" # end _sql
from google.colab.sql import bigquery as _bqsqlcell
df = _bqsqlcell.run(_sql)
df
```
